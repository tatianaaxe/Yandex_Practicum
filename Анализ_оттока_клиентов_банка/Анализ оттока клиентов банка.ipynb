### Подготовка данных:
- форматирование стиля столбцов, корректировка формата данных, удаление дубликатов, анализ гистограмм;
- анализ природы пропусков (возраст, баланс) и их разумная обработка (создать новый признак, принимающий 1, если в наблюдении пропущено значение признака, и 0, если в наблюдении значение признака не пропущено. Далее этот признак использовать как столбец сводной таблицы, где по каждому признаку будет посчитана мера центральной тендеции . Цель такого анализа - найти зависимости пропусков от данных или подтвердить отсутствие таких зависимостей.);  
- кодирование (для использования в анализе корреляции).


### EDA:
- анализ распределения признаков, в тч в разрезе оттока (непрерывные - ищем отточные интервалы, категориальные/дискретные - ищем значения признака с оттоком выше, чем по банку):  
1. Непрерывные признаки: их мы анализируем, чтобы найти интервалы признака, на которых отток выше, чем лояльность;
2. Категориальные/дискретные признаки: их мы анализируем, чтобы понять, какое из значений категориального признака более или менее склонно к оттоку;
- анализ портретов клиентов в разрезе оттока; 
- корреляционный анализ (в тч интерпретация).
- портрет клиента (меры центральной тендецнии, описывающие портреты клиентов, собрать в виде таблицы, и по каждому признаку между сегментами будет посчитана относительная разница);
        
   
### Проверка статистических гипотез:
- различия дохода между теми клиентами, которые ушли и теми, которые остались
- самостоятельная гипотеза (зависимость оттока от пола, возраста клиентов, города).
- обоснование выбора теста (ЦПТ, дисперсии,бинарность значений)


### Подведение промежуточных итогов  
- выводы по результатам EDA, корреляционного анализа и гипотез, выделение признаков, влияющих на оттток.

На этом этапе выбрать признаки, которые лягут в основу сегментации, а также перечислить их - выбор нужно делать по результатам EDA, корреляции и гипотез, при этом признаки и их значения/интервалы должны соответствовать выводам из EDA в разрезе оттока.
        
Не стоит выделять много сегментирующих признаков, чтобы сегменты не получились слишком узкими - так мы не сможем охватывать много клиентов в пределах сегмента, и это повышает риск того, что мероприятие по удержанию клиентов не окажется эффективным.

```python
# поправим стиль столбцов
data.columns = data.columns.str.lower()
data = data.rename(columns = {'userid':'user_id'})
```

Кажется странным, что с нулевым балансом только 2 записи - есть вероятность, что пропуски это и есть нулевой баланс.

Для столбца balance необходимо создать признак, принимающий 1, если пропущено значение и 0, если значение не пропущено -  чтобы найти зависимости пропусков от данных или их отсутствие.    
Также нужно закодировать столбцы с категориальными неупорядоченными значениями - gender и city.

```python
data['balance_nan'] = np.where(data['balance'].isna(), 1, 0)
data['age_nan'] = np.where(data['age'].isna(), 1, 0)
```

```python
data_copy = data.copy()
data = pd.get_dummies(data, dummy_na=True, columns=['gender', 'city'])
data['city'] = data_copy['city']
data['gender'] = data_copy['gender']

data.head()
```

```python
data.drop(['gender_nan', 'city_nan'], axis= 1 , inplace= True)
data.head()
```

```python
# посмотрим на корреляцию данных по методу Спирмена с оттоком
 
data_corr_spearman =  data.iloc[:, 1:].corrwith(data['age_nan'], method = 'spearman')
data_corr_spearman_mx = pd.DataFrame(data=data_corr_spearman)


# отобразим heatmap
plt.figure(figsize = (15,15))
sns.heatmap(data_corr_spearman_mx, square=True, vmax=0.8, annot = True, cmap='coolwarm')

plt.show();
```

Корреляции пропусков в возрасте с другими значениями нет, можно отнести их к MAR.

```python
# посмотрим на корреляцию данных по методу Спирмена с оттоком
 
data_corr_spearman =  data.iloc[:, 1:].corrwith(data['balance_nan'], method = 'spearman')
data_corr_spearman_mx = pd.DataFrame(data=data_corr_spearman)


# отобразим heatmap
plt.figure(figsize = (15,15))
sns.heatmap(data_corr_spearman_mx, square=True, vmax=0.8, annot = True, cmap='coolwarm')

plt.show();
```

## Проверка статистических гипотез

```python
# зададим уровень статистической значимости
alpha = .05 
```

### Гипотеза №1

H0: Нет различий в среднем доходе между клиентами, которые ушли и которые остались 

H1: Есть различия в среднем доходе между клиентами, которые ушли и которые остались

```python
# построим выборки по признаку оттока в разрезе дохода
no_churn_per_salary = (data
            .query('churn == 0')
            .pivot_table(index = 'user_id', values = 'est_salary')
                      )
churn_per_salary = (data
         .query('churn == 1')
         .pivot_table(index = 'user_id', values = 'est_salary')
                      )

print(no_churn_per_salary.describe())
print(churn_per_salary.describe())
```

```python
sns.histplot(data=no_churn_per_salary, x='est_salary')
plt.title('Распределение выборок')
plt.xlabel('Доход')                   
plt.ylabel('Количество');
sns.histplot(data=churn_per_salary, x='est_salary');
```

```python
# посчитаем дисперсию для получившихся выборок
disp_1 = np.var(no_churn_per_salary)
disp_2 = np.var(churn_per_salary)

print(f'Разница между дисперсиями: {disp_1[0]-disp_2[0]}')
```

Поскольку размеры выборок большие, распределение нормальное, а разница между дисперсиями и статистическими отклонениями достаточно велика, для проверки гипотез можно использовать критерий Стьюдента, но с указанием критерия Уэлча.

```python
results = (stats.ttest_ind(
    no_churn_per_salary, 
    churn_per_salary,
    equal_var=False)
          )
    
print('p-значение: ', results.pvalue)
    
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Гипотетически, нет различий в среднем доходе между клиентами, которые ушли и которые остались.

```python
from scipy.stats import shapiro
shapiro(no_churn_per_salary)
```

```python
shapiro(churn_per_salary)
```

По тесту Шапиро, выборки не имеют нормального распределения (pvalue=0.0), поэтому воспользуемся критерием Манна-Уитни.

```python
results_MW = stats.mannwhitneyu(no_churn_per_salary, churn_per_salary)

print('p-значение: ', results_MW.pvalue)
    
if (results_MW.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

По критерию Манна-Уитни, есть различия в среднем доходе между клиентами, которые ушли и которые остались.

### Гипотеза №2
H0: Нет различий в ушедших клиентах из Ярославля и из Ростова

H1: Есть различия в ушедших клиентах из Ярославля и из Ростова

Поскольку показатели бинарные, необходимо использовать Z-тест пропорций.

```python
from statsmodels.stats.proportion import proportions_ztest
# построим выборки в разрезе двух городов 

churn_per_Ярославль = data.query('city_Ярославль == 1')
            
                      
churn_per_Рыбинск = data.query('city_Рыбинск == 1')
         

# количество отточных
count = np.array([
    len(churn_per_Ярославль[churn_per_Ярославль['churn']==1]),
    len(churn_per_Рыбинск[churn_per_Рыбинск['churn']==1])
])
# общее количество
nobs = np.array([len(churn_per_Ярославль), len(churn_per_Рыбинск)])
# сформириуем Z-тест для пропорций
z_stat, p_value = proportions_ztest(count, nobs)

print(f'Z-statistic = {z_stat:.3f}; p-значение:  {p_value:.3f}')
```

```python
churn_per_Ярославль['churn']==1
```

```python
print(f'p-значение: {round(p_value,3)}')
if (p_value < alpha):
    print("Отвергаем нулевую гипотезу: между долями есть значимая разница")
else:
    print("Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными")
```

Гипотетически, есть различия между ушедшими клиентами из Ярославля и Рыбинска.
